```python
# Langkah 2: Import library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from IPython.display import display # Tambahkan import untuk display

# Langkah 3: Baca data dari file jaringan
X = pd.read_csv("/content/drive/My Drive/Machine Learning/Clustering/dataset/network_logs.csv")
X = X[["Length", "Protocol"]]  # Pilih kolom yang relevan untuk clustering

# Langkah 4: Tampilkan data
display(X)
```

```python
# Langkah 5: Inisialisasi dan latih model K-Means
model = KMeans(n_clusters=2, random_state=42)
model.fit(X)

# Langkah 6: Tambahkan hasil cluster ke DataFrame
X["cluster"] = model.labels_
print("Clustering berhasil dan label cluster telah ditambahkan ke DataFrame.") # Komentar jika berhasil
```

```python
# Langkah 7: Visualisasi hasil clustering
plt.scatter(X["Length"], X["Protocol"], c=X["cluster"], cmap="plasma")
plt.title("Hasil Clustering Lalu Lintas Jaringan")
plt.xlabel("Length")
plt.ylabel("Protocol")
plt.grid(True)
plt.show()
```

```python
plt.scatter(X["Length"], X["Protocol"], c=X["cluster"], cmap="plasma")
# Get the cluster centers from the KMeans model and assign them to centroids
centroids = model.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroid')
# The following lines were also indented, likely by mistake.
# They should be aligned with the previous two plt.scatter calls
plt.title("Hasil Clustering Lalu Lintas Jaringan")
plt.xlabel("Length")
plt.ylabel("Protocol")
plt.grid(True)
plt.legend()
plt.show()
```
